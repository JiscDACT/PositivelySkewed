## Return Values and Return Statements

The **return** keyword allows you to exit a function and return a value. You can assign a function to a variable, then use it to store a result until you need it. You might do this if you want to use the result of one function in a future function, instead of printing it to the terminal.

In the code example below, the variable **sum** calls the function "addition(number1, number2)" and stores the result because of the line "return a+b". The result of the function is then output when the variable "sum" is printed. 

````python
# The function addition takes two variables and adds them together when called
def addition(a, b):
  return a+b

# Variables to be added
number1 = 5
number2 = 4

# Variable that calls the function and stores returned restult
sum = addition(number1, number2)

# Get the stored result of the function
print(sum)
````

In the code example below, printing the variable "minus" calls the function "subtraction(sum, number3)". When the varible "sum" is passed, it calls the function "addition(number1, number2)", and stores the result because of the line "return a+b". The result of the "addition()" function is then passed back to the function "subtraction(sum, number3)". The result of the "subtraction()" function is stored in variable "minus" due to the line "return c-d", and is then output to the terminal when the variable minus is printed.

````python
# Variables to be used
number1 = 5
number2 = 4
number3 = 18

# The function addition takes two arguments and adds them together
def addition(a, b):
  return a+b

# The function subtraction() takes two arguments and subtracts them
def subtraction(c, d):
  return c-d

# The variable sum stores the returned result from the addition() function
sum = addition(number1, number2)

# The variable minus stores the returned result from the subtraction() 
# function. The result from the addition() function is passed to the 
# subtraction() function using the variable sum
minus = subtraction(sum, number3)

# The print statement prints the variable minus, which calls the subtraction()
# function, and prints the result 
print(minus)
````

# Loops

There are two types of loops in python:

* while loops
* for loops


### While Loops

The while loop executes code statements, as long as a condition is met. 

Conditions in python are made using variables, booleon types (TRUE or FALSE)  [standard logical operators](https://www.w3schools.com/python/python_operators.asp) (i.e., !=, >=, < etc), and, or, and not.

The image below breaks down an example of a while loop, marking each step with a red number:

1. We create a variable that will be part of our condition
2. We use the keyword "while" to create the while loop
3. We create a condition (in the red circle) and end it with a colon. The condition is important because the code will only repeat if the condition is being met. In this example, the code will repeat as long as the variable i is less than the value 6
4. As long as the condition is met, the indented code after the colon will be executed

image.png


For example: "while" a variable is less than 6, repeat these actions. 

Run the code below to see the output. In this example, we increment the value of "i" by 1 (using the code "i+=1"), this is to prevent the loop infinitely repeating as otherwise "i" will always be less than 6

````python
# Condition variable
i = 1

# While loop
while i < 6:
  print(i) # Print the value of i
  i+=1 # Increase the Value of i to prevent an infinite loop
````

* The variable "i" is used as a condition variable. It allows the loop to repeat
* The keyword "while" creates a loop
* "i < 6" creates a condition. The loop will repeat untill the condition stops being true
* "Print(i)" is our statement to be executed as long as the loop repeats
* "i+=1" increments the condition variable, so eventually "i" will equal 6 and the loop will stop. (Forgetting this will cause an infinite loop!) 

#### Else Statement

The else statement runs code when the condition in the while loop is no longer being met. 

Using the same code example from above, an else statement has been added at the end using the keyword "else" followed by a colon. The indented code below the else statement only runs when the while loop has ended. You can see the output by running the code.

````python
# Condition variable
i = 0

# While loop
while i < 6:
  print(i) # Print the value of a
  i+=1
else:
  print("\n i is greater than 5")
````

## For Loops

With the "for" keyword, a code block can be executed, one for each item in a list, dictionary etc. 

A "for loop" is structured by using the keywords "for" followed by a variable, and the keyword "in" followed by a sequence. The code block to be executed is then written below with an indentation.

````python
# Variable containing a list
DACT_teams = ["squAdge", "Beta Blockers", "Cheetah"]

# For loop
for teams in DACT_teams:
  print(teams) # Print the items in the list
````

In the code example above:

* The variable "DACT_teams" contains a list of three strings
* The keyword "for" is used to create the for loop
* The variable "teams" is a new variable, which holds the individual values of the list
* The keyword "in" and the variable "DACT_teams" identifies the list we want to loop through
* Print(teams) is our statement to be executed untill the loop is finished iterating through the list

````python
# Import packages
import seaborn as sns

# Load penguin dataset
df = sns.load_dataset('penguins')
data = df['species'].unique() # Filter the dataset to only unique species

# For loop to print each unique species in the data set
for penguin_species in data:
  print(penguin_species)
````

The code above shows an example of a for loop being used with a dataset:

* First we import the relevent package "seaborn"
* The dataset (called penguins - provided by the seaborn package) is assigned to the variable "df"
* We create a for loop with the keyword "for"
* The variable penguin_species is created, giving us access to each item in the list
* The keyword "in" is used to identify our list
* The list "df[species].unique()" is used. ("df[species]" tells the program we want all the items in the column 'species', ".unique()" tells the program to only output one row per unique item)
* The line "print(penguin_species)" prints each unique item in our list

# IF Statements

An If statement in python is used for decision making. There are three important statements:

* if
* elif
* else

## If

An if statment is written using the keyword "if", collowed by a condition (the same as a while loop), then a colon. When the condition of an if statement is met, the indented code is run. If the condition is not met, the program skips the indent. 

````python
# Variables
integer = 10

# If statement
if integer > 0:
  print('This is a positive number')
````

In the example above, our variable "integer" has the value 10. When the code is run, our if statement checks if our varialbe is greater than 0. This is "true", so our line "print('This is a positive number')" is run.

````python
# Variables
integer = 0

# If statement
if integer > 0 or integer == 0:
  print('This is a positive number')
````

In the code example above, we have used the "or" operator to be more specific in our if statement condition. Again, the indented code is executed because the condition has been met.

## Elif

The "elif" keyword is a way of providing an alternate condition in python. In otherwords, if the previous condition was not true, try this second condition. The elif statement must be written after an "if" statement, using the keyword "elif" followed by a condition, a colon, and indented code to be executed. 

````python
# Variables
integer = -10

# If statement
if integer > 0 or integer == 0:
  print('This is a positive number')
# Elif statement
elif integer < 0:
  print('This is a negative number')
````

In the code example above, our first if statement was not true, and so the acompanying indented code was not run. Python therefore checks to see if the second condition in the "elif" statement is true, which it is. The relevent indented code is then executed as the condition is met.

Note: If the variable "integer" had been a positive number, then the if statement and relevent indented code would have been run, and the "elif" statement would have been ignored.

## Else

The "else" keyword is a way of catching all other options not covered by the "if" and "elif" conditions. It is written using the keyword "else", followed by a colon and indented code to be executed. 

````python
# Variables
time = 21

# If statement
if time < 12:
  print('Good Morning')
# Elif statement
elif time < 17:
  print('Good Afternoon')
# Else statement
else:
  print('Good Evening')
````

In the code example above, the variable "time" is set to 21. As the variable is greater than 12 and 17, the "if" and "elif" conditions are ignored, and our variable is caught by the "else" statement.

## Nested IF Statements

You can also have if statements inside of other if statements! The structure is the same, except the second if statemet is indented to be within the first. The first if statement must be true for the second if statement to be tested. The code below shows this in practice.

````python
# Variable
number = 16.8

# First if statement
if number > 10:
  print('This number is greater than 10')
  # Second if statement
  if number < 20:
    print('This number is less than 20')
  # Else statement to catch everything else
  else:
    print('This number is greater than 20')
````

If statements can also be nested in other structures, such as for loops. The code example below shows an example of a an if statement nested inside of a for loop.

````python
# Import packages
import seaborn as sns

# Load dataset
df = sns.load_dataset('penguins')
# Filter dataset
mass = df['body_mass_g']

# For loop
for value in mass:
  # Nested if statement
  if value > 4000:  # If condition
    print('That is a large penguin')
  else: 
    print('This penguin is a standard weight')
````

In the code above, we are using the penguin data set, and filtering it to only contain informaiton on the mass of each penguin. 

* Using a "for" loop, we obtain the weight of every penguin using the variable "value" and specifing our dataset "mass". 
* We then use our nested if statement to specify conditions. 
* If the "value" of a penguins "mass" is "greater than 4000" then our if statement is true and we print the relevent code "print('This is a large penguin')"
* If the condition is not met, then the "else" statement catches this, and executes the line "print('This penguin is a standard weight')"
